```python
!pip install gTTS
#import Packages
from gtts import gTTS #Google Text To Speech
```

```python
#saving 'hello' in English to audio file
tts=gTTS('hello',lang='en')
tts.save('./hello.mp3')
```

```python
#'hello bonjour adios' in English & French & Spanish respectively to 'hello_bonjour_adios' mp3 file
tts_en=gTTS('hello',lang='en')
tts_fr=gTTS('bonjour',lang='fr')
tts_sp=gTTS('adios',lang='es')
with open('./hello_bonjour_adios.mp3','wb') as f:
    tts_en.write_to_fp(f)
    tts_fr.write_to_fp(f)
    tts_sp.write_to_fp(f)
```
