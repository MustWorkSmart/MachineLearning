```python
# Importing necessary library
import pandas as pa
import numpy as np
import nltk
#nltk.download('punkt_tab') #needed to avoid this when using nltk for the first time:
'''
LookupError: 
**********************************************************************
  Resource punkt_tab not found.
  Please use the NLTK Downloader to obtain the resource:

  >>> import nltk
  >>> nltk.download('punkt_tab')
'''
import os
import nltk.corpus

# sample text for performing tokenization
text = "In Hong Kong we drive on the left-hand side of the road."

# importing word tokenize from nltk
from nltk.tokenize import word_tokenize

# Passing the string text into word tokenize for breaking the sentences
token = word_tokenize(text)
print(token) #note that left-hand is 1 token
```

```python
#finding frequency (distinct)
from nltk.probability import FreqDist
fdist = FreqDist(token)
print(fdist)
```

```python
fdist
```

```python
#to find the freq of top N words
N = 5
fdist_topN = fdist.most_common(N)
print(fdist_topN)
```

```python
#stemming - normalizing words into its base form or root form
# porterstemmer from nltkmport
from nltk.stem import PorterStemmer
pst = PorterStemmer()
pst.stem("thinking")
```

```python
#checking
stm = ["thought", "thinking", "thinks"]
for word in stm:
    print(word + ":" + pst.stem(word))
stm2 = ["giving", "given", "gave", "give"]
for word in stm2:
    print(word + ":" + pst.stem(word))
```

```python
#another stemmer - seems to be more agressive than PorterStemmer above - note 3 cases of giv below
from nltk.stem import LancasterStemmer
lst = LancasterStemmer()
for word in stm:
    print(word + ":" + lst.stem(word))
for word in stm2:
    print(word + ":" + lst.stem(word))
```

```python
#Lemmatization - convert a word to its base form
#as compared to stemming, Lemmatization considers the context and converts the word to its meaningful base form
#while stemming above just removes the last few char's, often leading to incorrect meanings and spelling errors
```

```python
#stop words - most common words like a, at, the, for, is, on, in
#these words do not provide any meaning are are usually removed from texts
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwds = set(stopwords.words('english'))
```

```python
text_lower = word_tokenize(text.lower())
print(text_lower)
non_stopwords = [x for x in text_lower if x not in stopwds]
print(non_stopwords)
```
